## definir melhor quantidade de clusters para o k-mean

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Dados de exemplo
data = bank[['age', 'job', 'marital', 'education', 'default']]

df = pd.DataFrame(data)

# Codificação one-hot para as colunas categóricas
df_encoded = pd.get_dummies(df[['job', 'marital', 'education', 'default']])

# Lista para armazenar os valores de SSE
sse = []

# Valores de K a serem testados
k_values = range(1, 11)

# Executar K-means para diferentes valores de K
for k in k_values:
    kmeans = KMeans(n_clusters=k, init='random')
    kmeans.fit(df_encoded)
    sse.append(kmeans.inertia_)

# Plotar o gráfico do método do Elbow
plt.plot(k_values, sse, 'bx-')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('Soma dos Quadrados dos Erros (SSE)')
plt.title('Método do Elbow')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Lista para armazenar os valores de Silhouette Score
silhouette_scores = []

# Valores de K a serem testados
k_values = range(4, 9)

# Executar K-means para diferentes valores de K
for k in k_values:
    kmeans = KMeans(n_clusters=k, init='random')
    kmeans.fit(df_encoded)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(df_encoded, labels))

# Imprimir os resultados do Silhouette Score para cada valor de K
for k, score in zip(k_values, silhouette_scores):
    print(f"K = {k}: Silhouette Score = {score}")
```

## teste inicial k-mean

```python
import pandas as pd
from sklearn.cluster import KMeans

# Dados de exemplo
dataa = bank[['age', 'job', 'marital', 'education', 'default']]

df = pd.DataFrame(dataa)

# Codificação one-hot para as colunas categóricas
df_encoded = pd.get_dummies(df[['job', 'marital', 'education', 'default']])

# Passo 1: Inicialização aleatória dos centróides
k = 3
kmeans = KMeans(n_clusters=k, init='random')
kmeans.fit(df_encoded)

# Passo 2: Atribuição dos pontos aos centróides mais próximos
df['Cluster'] = kmeans.predict(df_encoded)

# Exibir tabela com os dados e atribuição dos clusters
df
```

## Agrupando por cluster

```python
# Agrupar por cluster e calcular a média das idades
idade_media = df.groupby('Cluster')['age'].mean()

# Calcular as strings mais frequentes nas outras colunas
strings_frequentes = df.groupby('Cluster').agg(lambda x: x.mode().iloc[0])[['job', 'marital', 'education', 'default']]

# Adicionar uma coluna para contar a quantidade de linhas em cada cluster
quantidade_linhas = df.groupby('Cluster').size()

# Combinar os resultados em uma única tabela
testee = pd.concat([idade_media, strings_frequentes, quantidade_linhas.rename('Quantidade de Linhas')], axis=1)
testee
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Selecionando as colunas desejadas
selected_columns = ['age', 'job', 'marital', 'education', 'default', 'housing', 'loan', 'poutcome', 'y']
data = bank_final[selected_columns]

# Padronizando a variável 'age'
scaler = StandardScaler()
data['age'] = scaler.fit_transform(data['age'].values.reshape(-1, 1))

# Codificação one-hot das variáveis categóricas
categorical_columns = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'poutcome', 'y']
encoded_data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)

# Aplicando o PCA
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(encoded_data)

# Variância explicada pelos componentes principais
explained_variance_ratio = pca.explained_variance_ratio_

# Plotando gráfico de dispersão dos componentes principais
plt.figure(figsize=(8, 6))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Gráfico de Dispersão - Componentes Principais')
plt.show()

# Plotando a variância explicada acumulada
cumulative_variance = np.cumsum(explained_variance_ratio)
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel('Número de Componentes Principais')
plt.ylabel('Variância Explicada Acumulada')
plt.title('Variância Explicada Acumulada pelo PCA')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Executando o algoritmo K-means com K = 3
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(transformed_data)

# Adicionando os rótulos de cluster aos dados
clustered_data = pd.DataFrame(transformed_data, columns=['Componente Principal 1', 'Componente Principal 2'])
clustered_data['Cluster'] = cluster_labels

# Tabela com a média das variáveis para cada cluster
cluster_means = clustered_data.groupby('Cluster').mean()
print(cluster_means)

# Gráfico de dispersão com os clusters coloridos
plt.figure(figsize=(8, 6))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], c=cluster_labels, cmap='viridis')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Gráfico de Dispersão - Segmentação de Clientes')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Calcular o número de ocorrências em que y é igual a 'yes'
y_yes = (bank_ad['y'] == 'yes').sum()

# Calcular a taxa de sucesso
taxa = y_yes / len(bank_ad['y'])

# Agrupar a contagem de conversões por mês
conversion_by_month = bank_ad[bank_ad['y'] == 'yes'].groupby('month').size()

# Calcular a taxa de sucesso por mês
taxa_by_month = (conversion_by_month / len(bank_ad['y'])) * 100

# Criar o gráfico de barras
taxa_by_month.plot(kind='bar')

# Definir os rótulos e o título do gráfico
plt.xlabel('Month')
plt.ylabel('Conversion Rate (%)')
plt.title('Conversion Rate by Month')

# Exibir o gráfico
plt.show()
```
